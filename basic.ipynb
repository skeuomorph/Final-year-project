```python
from causalgraphicalmodels import CausalGraphicalModel

sect = CausalGraphicalModel(
    nodes=["officer_defined_ethnicity", "stopped_and_searched", "charged"],
    edges=[
        ("officer_defined_ethnicity", "stopped_and_searched"), 
        ("stopped_and_searched", "charged")
    ]
)

sect.draw()
```

```python
print(sect.get_distribution())
```

```python
from causalgraphicalmodels import StructuralCausalModel
import numpy as np

#PLACEHOLDER FUNCTIONS FOR NOW
scm = StructuralCausalModel({
    "o_d_ethnicity": lambda     
            n_samples: np.random.binomial(n=1,p=0.7,size=n_samples),
    
    "s_and_s": lambda o_d_ethnicity, 
            n_samples: np.random.normal(loc=o_d_ethnicity, scale=0.1),
    
    "charged": lambda s_and_s, 
            n_samples: s_and_s ** 2,
})
```

```python
ds = scm.sample(n_samples=100)

ds.head()
```
